```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("HANDWRITTEN CHARACTER RECOGNITION - CNN PROJECT")
print("="*70)

# ============= STEP 1: LOAD AND EXPLORE DATASET =============
print("\n1. LOADING MNIST DATASET")
print("-" * 70)

# Load MNIST dataset (handwritten digits 0-9)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"✓ MNIST Dataset loaded successfully!")
print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
print(f"Number of classes: {len(np.unique(y_train))}")
print(f"Classes: {np.unique(y_train)}")

print(f"\nClass Distribution in Training Set:")
unique, counts = np.unique(y_train, return_counts=True)
for digit, count in zip(unique, counts):
    print(f"  Digit {digit}: {count} samples")

# ============= STEP 2: DATA PREPROCESSING =============
print("\n" + "="*70)
print("2. DATA PREPROCESSING")
print("-" * 70)

# Normalize pixel values to 0-1 range
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

print(f"✓ Pixel values normalized to [0, 1]")

# Reshape data for CNN (add channel dimension)
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

print(f"✓ Data reshaped for CNN")
print(f"Training shape: {X_train.shape}")
print(f"Testing shape: {X_test.shape}")

# Convert labels to categorical
y_train_cat = keras.utils.to_categorical(y_train, 10)
y_test_cat = keras.utils.to_categorical(y_test, 10)

print(f"✓ Labels converted to categorical format")

# ============= STEP 3: DATA VISUALIZATION =============
print("\n" + "="*70)
print("3. DATA VISUALIZATION")
print("-" * 70)

# Visualize sample images
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
fig.suptitle('Sample Handwritten Digits from MNIST Dataset', fontsize=14, fontweight='bold')

for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[i].reshape(28, 28), cmap='gray')
    ax.set_title(f'Digit: {y_train[i]}', fontweight='bold')
    ax.axis('off')

plt.tight_layout()
plt.savefig('01_sample_digits.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 01_sample_digits.png")
plt.show()

# ============= STEP 4: BUILD CNN MODEL =============
print("\n" + "="*70)
print("4. BUILDING CNN MODEL")
print("-" * 70)

model = models.Sequential([
    # First Convolutional Block
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Second Convolutional Block
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Third Convolutional Block
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.25),

    # Flatten and Dense Layers
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    layers.Dense(128, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    # Output Layer
    layers.Dense(10, activation='softmax')
])

print("✓ CNN Model Architecture:")
print("-" * 70)
model.summary()

# Compile model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\n✓ Model compiled successfully!")

# ============= STEP 5: TRAIN MODEL =============
print("\n" + "="*70)
print("5. TRAINING MODEL")
print("-" * 70)

# Data augmentation
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    fill_mode='nearest'
)

print("Training the CNN model...")
print("This may take a few minutes...\n")

history = model.fit(
    datagen.flow(X_train, y_train_cat, batch_size=128),
    epochs=20,
    validation_data=(X_test, y_test_cat),
    steps_per_epoch=len(X_train) // 128,
    verbose=1
)

print("\n✓ Model training completed!")

# ============= STEP 6: EVALUATE MODEL =============
print("\n" + "="*70)
print("6. MODEL EVALUATION")
print("-" * 70)

# Get predictions
y_pred_proba = model.predict(X_test)
y_pred = np.argmax(y_pred_proba, axis=1)

# Calculate metrics
test_loss, test_accuracy = model.evaluate(X_test, y_test_cat, verbose=0)

print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

print(f"\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))

print(f"\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
print(cm)

# ============= STEP 7: VISUALIZATIONS =============
print("\n" + "="*70)
print("7. GENERATING VISUALIZATIONS")
print("-" * 70)

# Plot 1: Training History - Loss
print("\n1. Generating Training History - Loss...")
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(history.history['loss'], label='Training Loss', linewidth=2, marker='o', markersize=4)
ax.plot(history.history['val_loss'], label='Validation Loss', linewidth=2, marker='s', markersize=4)
ax.set_xlabel('Epoch', fontweight='bold', fontsize=12)
ax.set_ylabel('Loss', fontweight='bold', fontsize=12)
ax.set_title('Model Loss over Epochs', fontweight='bold', fontsize=14)
ax.legend(fontsize=11)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('02_training_loss.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 02_training_loss.png")
plt.show()

# Plot 2: Training History - Accuracy
print("\n2. Generating Training History - Accuracy...")
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2, marker='o', markersize=4)
ax.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2, marker='s', markersize=4)
ax.set_xlabel('Epoch', fontweight='bold', fontsize=12)
ax.set_ylabel('Accuracy', fontweight='bold', fontsize=12)
ax.set_title('Model Accuracy over Epochs', fontweight='bold', fontsize=14)
ax.legend(fontsize=11)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('03_training_accuracy.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 03_training_accuracy.png")
plt.show()

# Plot 3: Confusion Matrix
print("\n3. Generating Confusion Matrix...")
fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
            cbar_kws={'label': 'Count'}, annot_kws={'size': 10, 'weight': 'bold'})
ax.set_xlabel('Predicted Label', fontweight='bold', fontsize=12)
ax.set_ylabel('True Label', fontweight='bold', fontsize=12)
ax.set_title('Confusion Matrix - CNN Model', fontweight='bold', fontsize=14)

plt.tight_layout()
plt.savefig('04_confusion_matrix.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 04_confusion_matrix.png")
plt.show()

# Plot 4: Per-Class Accuracy
print("\n4. Generating Per-Class Accuracy...")
per_class_accuracy = []

for digit in range(10):
    mask = y_test == digit
    acc = accuracy_score(y_test[mask], y_pred[mask])
    per_class_accuracy.append(acc)

fig, ax = plt.subplots(figsize=(10, 6))

colors = plt.cm.viridis(np.linspace(0, 1, 10))
bars = ax.bar(range(10), per_class_accuracy, color=colors, edgecolor='black', linewidth=1.5)

ax.set_xlabel('Digit Class', fontweight='bold', fontsize=12)
ax.set_ylabel('Accuracy', fontweight='bold', fontsize=12)
ax.set_title('Per-Class Accuracy', fontweight='bold', fontsize=14)
ax.set_xticks(range(10))
ax.set_ylim([0, 1.05])
ax.grid(axis='y', alpha=0.3)

# Add value labels
for bar, acc in zip(bars, per_class_accuracy):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.02,
            f'{acc:.3f}', ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.savefig('05_per_class_accuracy.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 05_per_class_accuracy.png")
plt.show()

# Plot 5: Correctly and Incorrectly Classified Digits
print("\n5. Generating Classification Examples...")
fig, axes = plt.subplots(2, 5, figsize=(14, 6)) # Changed from 3x5 to 2x5
fig.suptitle('Correctly and Incorrectly Classified Examples', fontsize=14, fontweight='bold')

# Get correctly classified
correct_indices = np.where(y_pred == y_test)[0]
incorrect_indices = np.where(y_pred != y_test)[0]

# Show correct classifications (first 5)
for i, idx in enumerate(correct_indices[:5]): # Changed from 7 to 5
    ax = axes[0, i]
    ax.imshow(X_test[idx].reshape(28, 28), cmap='gray')
    ax.set_title(f'True: {y_test[idx]}, Pred: {y_pred[idx]}',
                 fontweight='bold', color='green', fontsize=10)
    ax.axis('off')

# Show incorrect classifications (first 5) # Changed from 8 to 5
for i, idx in enumerate(incorrect_indices[:5]):
    ax = axes[1, i] # Changed row index from dynamic to 1
    ax.imshow(X_test[idx].reshape(28, 28), cmap='gray')
    ax.set_title(f'True: {y_test[idx]}, Pred: {y_pred[idx]}',
                 fontweight='bold', color='red', fontsize=10)
    ax.axis('off')

# Hide empty subplots - no longer needed with 2x5 grid and 5 examples
# for i in range(len(incorrect_indices[:8]), 10):
#     axes[2, i % 5].axis('off')

plt.tight_layout()
plt.savefig('06_classification_examples.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 06_classification_examples.png")
plt.show()

# Plot 6: Model Performance Summary
print("\n6. Generating Model Performance Summary...")
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Model Performance Summary', fontsize=14, fontweight='bold')

# Plot 1: Final Accuracy
ax = axes[0, 0]
ax.barh(['Test Accuracy'], [test_accuracy], color='#3498db', edgecolor='black', linewidth=2)
ax.set_xlim([0, 1])
ax.set_title('Overall Test Accuracy', fontweight='bold')
ax.text(test_accuracy + 0.02, 0, f'{test_accuracy:.4f}', va='center', fontweight='bold', fontsize=12)
ax.grid(axis='x', alpha=0.3)

# Plot 2: Precision, Recall, F1-Score per class (average)
ax = axes[0, 1]
report_dict = classification_report(y_test, y_pred, output_dict=True)
precision_scores = [report_dict[str(i)]['precision'] for i in range(10)]
recall_scores = [report_dict[str(i)]['recall'] for i in range(10)]
f1_scores = [report_dict[str(i)]['f1-score'] for i in range(10)]

ax.plot(range(10), precision_scores, marker='o', label='Precision', linewidth=2, markersize=6)
ax.plot(range(10), recall_scores, marker='s', label='Recall', linewidth=2, markersize=6)
ax.plot(range(10), f1_scores, marker='^', label='F1-Score', linewidth=2, markersize=6)
ax.set_xlabel('Digit Class', fontweight='bold')
ax.set_ylabel('Score', fontweight='bold')
ax.set_title('Precision, Recall, F1-Score per Digit', fontweight='bold')
ax.set_xticks(range(10))
ax.legend()
ax.grid(alpha=0.3)

# Plot 3: Loss Comparison
ax = axes[1, 0]
ax.bar(['Training Loss', 'Test Loss'],
       [history.history['loss'][-1], test_loss],
       color=['#e74c3c', '#f39c12'], edgecolor='black', linewidth=2)
ax.set_ylabel('Loss', fontweight='bold')
ax.set_title('Final Training vs Test Loss', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# Plot 4: Epoch-wise improvement
ax = axes[1, 1]
epochs_range = range(1, len(history.history['accuracy']) + 1)
ax.fill_between(epochs_range, history.history['accuracy'], alpha=0.3, label='Training')
ax.fill_between(epochs_range, history.history['val_accuracy'], alpha=0.3, label='Validation')
ax.plot(epochs_range, history.history['accuracy'], marker='o', linewidth=2, markersize=4)
ax.plot(epochs_range, history.history['val_accuracy'], marker='s', linewidth=2, markersize=4)
ax.set_xlabel('Epoch', fontweight='bold')
ax.set_ylabel('Accuracy', fontweight='bold')
ax.set_title('Accuracy Improvement Over Epochs', fontweight='bold')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('07_performance_summary.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 07_performance_summary.png")
plt.show()

# ============= STEP 8: SAVE MODEL =============
print("\n" + "="*70)
print("8. SAVING MODEL")
print("-" * 70)

model.save('handwritten_recognition_model.h5')
print("✓ Model saved as: handwritten_recognition_model.h5")

# ============= STEP 9: TEST ON CUSTOM PREDICTION =============
print("\n" + "="*70)
print("9. TESTING MODEL ON SAMPLE IMAGES")
print("-" * 70)

# Predict on random test samples
sample_indices = np.random.choice(len(X_test), 10, replace=False)

fig, axes = plt.subplots(2, 5, figsize=(14, 6))
fig.suptitle('Model Predictions on Random Test Samples', fontsize=14, fontweight='bold')

for idx, ax in enumerate(axes.flat):
    test_idx = sample_indices[idx]
    image = X_test[test_idx].reshape(28, 28)

    prediction = model.predict(X_test[test_idx:test_idx+1], verbose=0)
    pred_class = np.argmax(prediction)
    confidence = np.max(prediction)
    true_class = y_test[test_idx]

    color = 'green' if pred_class == true_class else 'red'

    ax.imshow(image, cmap='gray')
    ax.set_title(f'True: {true_class}, Pred: {pred_class}\nConf: {confidence:.2%}',
                fontweight='bold', color=color, fontsize=10)
    ax.axis('off')

plt.tight_layout()
plt.savefig('08_sample_predictions.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 08_sample_predictions.png")
plt.show()

print("\n" + "="*70)
print("✓ PROJECT COMPLETE!")
print("="*70)
print("\n📊 Generated Files:")
print("  1. 01_sample_digits.png - Sample handwritten digits")
print("  2. 02_training_loss.png - Loss over epochs")
print("  3. 03_training_accuracy.png - Accuracy over epochs")
print("  4. 04_confusion_matrix.png - Confusion matrix heatmap")
print("  5. 05_per_class_accuracy.png - Accuracy per digit class")
print("  6. 06_classification_examples.png - Correct vs incorrect predictions")
print("  7. 07_performance_summary.png - 4-panel performance summary")
print("  8. 08_sample_predictions.png - Model predictions on random samples")
print("  9. handwritten_recognition_model.h5 - Trained model")
print("\n" + "="*70)
```
